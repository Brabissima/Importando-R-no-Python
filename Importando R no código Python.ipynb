```python
import rpy2.robjects as ro
```

```python
r = ro.r
r['source']('teste_r.R')
filter_target = ro.globalenv['filter_target']
filter_target
```

```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np

df = load_iris()
df = pd.DataFrame(data=np.c_[df['data'], df['target']], 
                  columns=df['feature_names']+['target'])
```

```python
df.sample(5)
```

```python
#df.to_csv('iris.csv')
#df = pd.read_csv('iris.csv')
```

```python
from rpy2.robjects import pandas2ri
from rpy2.robjects.conversion import localconverter
```

```python
with localconverter(ro.default_converter + pandas2ri.converter):
    r_from_pd = ro.conversion.py2rpy(df)
r_from_pd
```

```python
type(r_from_pd)
```

```python
df_result_r = filter_target(r_from_pd, 1)
df_result_r
```

```python
df_result = pandas2ri.rpy2py(df_result_r)
df_result
```
